```python
import pandas as pd
import matplotlib.pyplot as plt

#load data from CSV
mlb_stats = pd.read_csv('MLB_Advanced_Statistics.csv')

#extract relevant columns
wrc_plus = mlb_stats['wRC+']
ops = mlb_stats['OPS']
team_names = mlb_stats['Team']

#create scatter plot
plt.scatter(ops, wrc_plus, c=range(len(team_names)), cmap='rainbow')

#add axis labels and title
plt.xlabel('OPS')
plt.ylabel('wRC+')
plt.title('OPS vs wRC+ for MLB Teams')

#add team names as labels for each data point
for i, txt in enumerate(team_names):
    plt.annotate(txt, (ops[i], wrc_plus[i]), ha='center', va='center')

#show plot
plt.show()
```

```python
from scipy.stats import pearsonr

corr, _ = pearsonr(mlb_stats["OPS"], mlb_stats["wRC+"])
print("Correlation coefficient between OPS and wRC+: {:.2f}".format(corr))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#load data from CSV
mlb_stats = pd.read_csv('MLB_Advanced_Statistics.csv')

#extract relevant columns
k_rate = mlb_stats['K%']
batting_avg = mlb_stats['AVG']
team_names = mlb_stats['Team']

#create scatter plot
plt.scatter(batting_avg, k_rate, c=range(len(team_names)), cmap='rainbow')

#add axis labels and title
plt.xlabel('Batting Average (AVG)')
plt.ylabel('Strikeout Rate (K%)')
plt.title('Batting Average vs Strikeout Rate for MLB Teams')

#add team names as labels for each data point
for i, txt in enumerate(team_names):
    plt.annotate(txt, (batting_avg[i], k_rate[i]), ha='center', va='center')

#get the x and y limits of the plot
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()

#draw vertical and horizontal lines that intersect in the middle of the plot
plt.plot([(x_min + x_max) / 2, (x_min + x_max) / 2], [y_min, y_max], 'k--', lw=1)
plt.plot([x_min, x_max], [(y_min + y_max) / 2, (y_min + y_max) / 2], 'k--', lw=1)

#show plot
plt.show()
```

```python
import csv

def get_team_stats(team_name):
    with open('MLB_Model.csv') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            if row['Team'] == team_name:
                hits = round(float(row['Hits']))
                walks = round(float(row['Walks']))
                total_bases = round(float(row['Total Bases']))
                stolen_bases = round(float(row['Stolen Bases']))
                at_bats = round(float(row['At Bats']))
                caught_stealing = round(float(row['Caught Stealing']))
                earned_runs = round(float(row['Earned Runs']))
                home_runs_allowed = round(float(row['Home Runs Allowed']))
                strikeouts = round(float(row['Strikeouts']))
                errors = round(float(row['Errors']))
                double_plays = round(float(row['Double Plays']))
                
                return hits, walks, total_bases, stolen_bases, at_bats, caught_stealing, earned_runs, home_runs_allowed, strikeouts, errors, double_plays
    
    #if the team was not found in the CSV file, raise exception
    raise ValueError("Team not found in CSV file")

#prompt user for input
home_team = input("Enter the home team: ")
away_team = input("Enter the away team: ")
location = input("Enter the location: ")

#get home team statistics from CSV file
hits_home, walks_home, total_bases_home, stolen_bases_home, at_bats_home, caught_stealing_home, earned_runs_home, home_runs_allowed_home, strikeouts_home, errors_home, double_plays_home = get_team_stats(home_team)

#get away team statistics from CSV file
hits_away, walks_away, total_bases_away, stolen_bases_away, at_bats_away, caught_stealing_away, earned_runs_away, home_runs_allowed_away, strikeouts_away, errors_away, double_plays_away = get_team_stats(away_team)

#calculate the Runs Created and Runs Allowed for the home and away teams
rc_home = ((hits_home + walks_home) * (total_bases_home + (0.55 * stolen_bases_home)) / (at_bats_home + walks_home)) - caught_stealing_home
rc_away = ((hits_away + walks_away) * (total_bases_away + (0.55 * stolen_bases_away)) / (at_bats_away + walks_away)) - caught_stealing_away
ra_home = (earned_runs_home + home_runs_allowed_home) * (1 - (1 / (1 + (strikeouts_home / 100)))) + errors_home + (0.55 * double_plays_home)
ra_away = (earned_runs_away + home_runs_allowed_away) * (1 - (1 / (1 + (strikeouts_away / 100)))) + errors_away + (0.55 * double_plays_away)
rs_home = rc_home
rs_away = rc_away

#compare the runs created and runs allowed for both teams to determine which team is likely to win
if rc_home > rc_away and rs_home > rs_away:
    winner = home_team
    final_score = f"{winner} wins with a score of {round(rs_home, 2)} to {round(rs_away, 2)}"
elif rc_home < rc_away and rs_home < rs_away:
    winner = away_team
    final_score = f"{winner} wins with a score of {round(rs_away, 2)} to {round(rs_home, 2)}"
else:
    final_score = "It's a tie!"

print(final_score)
```
